```python
df.columns = df.columns.str.strip()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce')
```

```python
if 'Returns' in df.columns:
    df['Returns'] = df['Returns'].fillna(0)
```

```python
print(df.isnull().sum())
```

```python
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Day'] = df['Order Date'].dt.day
```

```python
df['HighSales'] = df['Sales'].apply(lambda x: 'Yes' if x > 1000 else 'No')
df[['Sales','HighSales']].head(10)
```

```python
df['ShippingDays'] = (df['Ship Date'] - df['Order Date']).dt.days
df[['Order Date','Ship Date','ShippingDays']].head(10)
```

```python
df['ProfitMargin'] = (df['Profit'] / df['Sales']) * 100
df[['ProfitMargin','Sales','Profit']].head()
```

```python
df['OrderSize'] = df['Quantity'].apply(
    lambda x: 'Small' if x <= 2 else ('Medium' if x <= 5 else 'Large')
)
df[['OrderSize','Quantity']].head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

ax=sns.countplot(x='Category',data=df,color='red')
ax.bar_label(ax.containers[0])
plt.title("Number Of order per Category")
plt.show()
```

1)Office Supplies (3569) has the highest orders → dominates demand.


2)Furniture (1249) and Technology (1083) have much lower orders.


3)Office Supplies is about 3× higher than other categories.


4)Indicates frequent purchases (daily-use items).


5)Furniture & Technology are less frequent, high-value purchases.

```python
ax=sns.countplot(x='Region',data=df,color="green")
ax.bar_label(ax.containers[0])
plt.title("Number of Order By region")
plt.show()
```

1)West (1901) is the top-performing region, followed by East (1688), while Central (1381) shows moderate demand and South (931) has the lowest orders.


2)There is a significant gap, with West having nearly 2× orders of South, indicating uneven demand distribution.


3)West and East together drive the majority of sales, making them key regions for business performance.


4)South is the weakest region, requiring focused improvement strategies.


5)Overall, there is a strong opportunity to boost low-performing regions and balance growth across all areas.

```python
ax=sns.countplot(x='Segment',data=df,color="yellow")
ax.bar_label(ax.containers[0])
plt.title("Number of Order By Segment")
plt.show()
```

1)Consumer segment has the highest orders (2997) → top-performing segment


2)Corporate segment (1774) shows moderate performance
Home Office segment (1130) has the lowest demand


3)Consumer orders are ~2.6× higher than Home Office → large gap

```python
ax=sns.barplot(x='Category',y="Sales",data=df,color="Darkblue")
plt.title("Average sales By Category")
plt.show()
```

1) Technology has the highest average sales (around 430–450), which means products in this category are generally more expensive and generate higher revenue per order.


2) Furniture shows moderate average sales (around 350–370), indicating medium-value transactions.


3) Office Supplies has the lowest average sales (around 170–190), suggesting items are low-cost but likely purchased frequently.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
for cat in df['Category'].unique():
    subset = df[df['Category'] == cat]
    plt.scatter(
        subset['Sales'],
        subset['Profit'],
        label=cat,
        alpha=0.6 
    )
plt.axhline(0)
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Sales vs Profit by Category')
plt.legend()
plt.grid()

plt.show()
```

Sales vs Profit by Category – Insights


Most data points are concentrated at low to medium sales values, showing most transactions are small or moderate.


There is a positive relationship between sales and profit overall (higher sales generally give higher profit).


📌 Category-wise Observation


Technology (green) shows the highest profits and also some very high sales values → high earning potential.


Office Supplies (orange) has many points near zero and negative profit → indicates frequent low-profit or loss-making sales.


Furniture (blue) is mostly around low to medium profit → moderate performance.

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Month'] = df['Order Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Sales'].sum()
monthly_sales.index = monthly_sales.index.astype(str)
plt.figure()
plt.plot(monthly_sales.index, monthly_sales.values, marker='o',color='red')
max_value = monthly_sales.max()
max_index = monthly_sales.idxmax()
plt.scatter(max_index, max_value, s=120)
plt.text(max_index, max_value, f'{max_value:.0f}')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Monthly Sales Trend')
plt.xticks(rotation=90)

plt.show()
```

1) Sales fluctuate throughout the period, showing ups and downs over time.


2) The highest sales occur around early 2020, reaching the peak (~47,970).


3) There is a sharp drop after the peak, indicating unstable sales during that period.


4) The lowest sales are seen in late 2019, showing a weak phase.


5) After mid-2020, sales become more stable, mostly staying between 25,000–33,000.

```python
corr = df[['Sales','Profit','Quantity']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

1)Sales and profit have a moderate positive correlation (0.33), meaning profit tends to increase with sales but not strongly.


2)Sales and quantity show a weak positive correlation (0.20), indicating higher quantity does not strongly impact total sales.


3)Profit and quantity have a very weak correlation (0.075), suggesting quantity sold has minimal effect on profit.


4)Overall, no strong correlations exist among the variables, indicating weak linear relationships.


5)Profit is likely influenced more by pricing, costs, or discounts rather than just sales volume or quantity.
